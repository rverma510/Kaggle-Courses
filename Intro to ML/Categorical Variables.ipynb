```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
help(pd.read_csv)
```

```python
train = pd.read_csv('train.csv', index_col = 'Id')
test = pd.read_csv('test.csv', index_col = 'Id')
```

```python
train.head()
```

```python
train.dtypes.unique()
```

```python
y = train.SalePrice
X = train.drop('SalePrice', axis = 1)

X_train_full, X_val_full, y_train, y_val = train_test_split(X, y, test_size = 0.2, random_state = 0)

X_train_full.dropna(axis = 1, inplace = True)
X_val_full.dropna(axis = 1, inplace = True)

low_cardinality_cols = [cname for cname in X_train_full.columns if X_train_full[cname].nunique() < 10 and X_train_full[cname].dtype == 'object']
numerical_cols = [cname for cname in X_train_full.columns if X_train_full[cname].dtype in ['int64', 'float64']]

my_cols = low_cardinality_cols + numerical_cols
X_train = X_train_full[my_cols].copy()
X_val = X_val_full[my_cols].copy()
```

```python
X_train.head()
```

```python
s = (X_train.dtypes == 'object')
object_cols = list(s[s].index)

print('Categorical variables')
print(object_cols)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

def score_dataset(X_train, X_val, y_train = y_train, y_val = y_val):
    model = RandomForestRegressor(n_estimators = 100, random_state = 0)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    return mean_absolute_error(y_val, y_pred)
```

```python
drop_X_train = X_train.select_dtypes(exclude = ['object'])
drop_X_val = X_val.select_dtypes(exclude = ['object'])

print("MAE from Approach 1 (Drop Categorical variables):")
print(score_dataset(drop_X_train, drop_X_val))
```

```python
from sklearn.preprocessing import LabelEncoder

label_X_train = X_train.copy()
label_X_val = X_val.copy()

label_encoder = LabelEncoder()
for col in object_cols:
    label_encoder.fit(X[col])
    label_X_train[col] = label_encoder.transform(X_train[col])
    label_X_val[col] = label_encoder.transform(X_val[col])
    
print("MAE from Approach 2 (Label Encoding):")
print(score_dataset(label_X_train, label_X_val))
```

```python
from sklearn.preprocessing import OneHotEncoder

OHEncoder = OneHotEncoder(handle_unknown = 'ignore', sparse = False)
OH_cols_train = pd.DataFrame(OHEncoder.fit_transform(X_train[object_cols]))
OH_cols_val = pd.DataFrame(OHEncoder.transform(X_val[object_cols]))

OH_cols_train.index = X_train.index
OH_cols_val.index = X_val.index

num_X_train = X_train.drop(object_cols, axis = 1)
num_X_val = X_val.drop(object_cols, axis = 1)

OH_X_train = pd.concat([num_X_train, OH_cols_train], axis = 1)
OH_X_val = pd.concat([num_X_val, OH_cols_val], axis = 1)
print(OH_X_train.shape)

print("MAE from Approach 3 (One-Hot Endcoding):")
print(score_dataset(OH_X_train, OH_X_val))
```
